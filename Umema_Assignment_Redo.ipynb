```python
#Question 6: 
#1. Filter the dataset to remove all entries where the "Country Name" is ‘Euro Area’.
new_countries = df[df['Country Name'] != 'Euro Area']
print(new_countries)
```

```python
#2. Filter the dataset to include only countries with a count of 14 or more entries.
filtered_countries = new_countries.groupby('Country Name').filter(lambda x: len(x) >= 14)
print(filtered_countries)
```

```python
#3. Remove the columns "Attribute", "count", and "Indicator Name" from the dataset.
cleaned_data = filtered_countries.drop(columns=['Attribute', 'Indicator Name'])
print(cleaned_data)
```

```python
#Question 7:
#It is now preferable to reshape the data from wide to long form so that:
#1. The years are organized under a new column named "Year".
#2. The values are organized under a new column named "Value".
#3. Your DataFrame will now have the following four columns: "Country Name", "Indicator Code", "Year", and "Value".

year_columns = [str(year) for year in range(2010, 2024)]  
long_form_data = cleaned_data.melt(id_vars=['Country Name', 'Indicator Code'], 
                                     value_vars=year_columns, 
                                     var_name='Year', 
                                     value_name='Value')
print(long_form_data)
```

```python
#5. Convert the "Value" into float64 data type
#6. print data types of all four columns and confirm
print(long_form_data['Value'].unique())
long_form_data['Value'] = pd.to_numeric(long_form_data['Value'], errors='coerce')
print(long_form_data)
```

```python
#Question 8: Now again reshape the data from long to wide so that:
#1. Each unique indicators from column "Indicator Code" become separate columns.
#2. The values are organized under these columns.
wide_form_data = long_form_data.pivot_table(index=['Country Name', 'Year'], 
                                             columns='Indicator Code', 
                                             values='Value')
wide_form_data.reset_index(inplace=True)
print(wide_form_data)
```

```python
#3. Your DataFrame will now have the following 18 columns: "Country Name", , "Year", and N, N1, N2, N3, N4, N5, N6, N7, R, R1, R2, R3, R4, R5, R6, R7. "Refer to Table 1 for
#details about the abbreviations."
#4. Order the columns as written in 3 above.

indicator_mapping = {
    'NGDP_XDC': 'N',
    'NCP_XDC': 'N1',
    'NCGG_XDC': 'N2',
    'NFI_XDC': 'N3',
    'NINV_XDC': 'N4',
    'NX_XDC': 'N5',
    'NM_XDC': 'N6',
    'NGDP_R_XDC': 'R',
    'NCP_R_XDC': 'R1',
    'NCGG_R_XDC': 'R2',
    'NFI_R_XDC': 'R3',
    'NINV_R_XDC': 'R4',
    'NX_R_XDC': 'R5',
    'NM_R_XDC': 'R6',
    'NSDGDP_XDC': 'N7',
    'NSDGDP_R_XDC': 'R7' 
}
wide_form_data.rename(columns=indicator_mapping, inplace=True)
desired_columns = ['Country Name', 'Year', 'N', 'N1', 'N2', 'N3', 'N4', 'N5', 'N6', 'N7', 'R', 'R1', 'R2', 'R3', 'R4', 'R5', 'R6', 'R7']
wide_form_data = wide_form_data[desired_columns]
print(wide_form_data)
```

```python
#Question 9: Replace all NA values with 0 in the 16 columns of data type float64 in the dataset.
wide_form_data.fillna(0, inplace=True)
print(wide_form_data)
```

```python
#Question 10: "Refer to Table 1 for details about the abbreviations." and
#1. Calculate Equation 7,Equation 8, and Equation 9
#Equation 7: Nominal_GDP = (N1 + N2 + N3 + N4) + (N5 − N6) + N7
wide_form_data['Nominal_GDP'] = (
    wide_form_data[['N1', 'N2', 'N3', 'N4']].sum(axis=1) + 
    (wide_form_data['N5'] - wide_form_data['N6']) + 
    wide_form_data['N7']
)
print(wide_form_data[['Country Name', 'Year', 'Nominal_GDP']])

#Equation 8: Nominal_GDP_DIFF = N − Nominal_GDP
wide_form_data['Nominal_GDP_DIFF'] = wide_form_data['N'] - wide_form_data['Nominal_GDP']
print(wide_form_data[['Country Name', 'Year', 'Nominal_GDP', 'Nominal_GDP_DIFF']])

#Equation 9: Nominal_GDP_PERC_SD = (Nominal_GDP_DIFF/N)*100
wide_form_data['Nominal_GDP_PERC_SD'] = (wide_form_data['Nominal_GDP_DIFF'] / wide_form_data['N']) * 100
print(wide_form_data[['Country Name', 'Year', 'Nominal_GDP', 'Nominal_GDP_DIFF', 'Nominal_GDP_PERC_SD']])
```

```python
#Question 10:
#2. Calculate Equation 10,Equation 11, and Equation 12
#Equation 10: Real_GDP = (R1 + R2 + R3 + R4) + (R5 − R6) + R7
wide_form_data['Real_GDP'] = (
    wide_form_data[['R1', 'R2', 'R3', 'R4']].sum(axis=1) + 
    (wide_form_data['R5'] - wide_form_data['R6']) + 
    wide_form_data['R7']
)
print(wide_form_data[['Country Name', 'Year', 'Real_GDP']])

#Equation 11: Real_GDP_DIFF = R − Real_GDP
wide_form_data['Real_GDP_DIFF'] = wide_form_data['R'] - wide_form_data['Real_GDP']
print(wide_form_data[['Country Name', 'Year', 'Real_GDP', 'Real_GDP_DIFF']])

#Equation 12: Real_GDP_PERC_SD = (Real_GDP_DIFF/R)*100
wide_form_data['Real_GDP_PERC_SD'] = (wide_form_data['Real_GDP_DIFF'] / wide_form_data['R']) * 100
print(wide_form_data[['Country Name', 'Year', 'Real_GDP', 'Real_GDP_DIFF', 'Real_GDP_PERC_SD']])
```

```python
#Question 11: 
#1. Drop NAs from dataframe
wide_form_data.dropna(inplace=True)
print(wide_form_data)

#2. Filter if both "Real_GDP_PERC_SD" and "Nominal_GDP_PERC_SD" are not in range as specified in Equation 6 i.e. in between -1% to 1% inclusive.
filtered_data_new = wide_form_data[
    (wide_form_data['Real_GDP_PERC_SD'] < -1) | 
    (wide_form_data['Real_GDP_PERC_SD'] > 1) |
    (wide_form_data['Nominal_GDP_PERC_SD'] < -1) | 
    (wide_form_data['Nominal_GDP_PERC_SD'] > 1)
]
print(filtered_data_new)
```

```python
#Question 12: Prepare a balanced panel such that
#1. Each country must have data for years 2010 till 2023.

years = list(range(2010, 2024))
country_years_count = wide_form_data.groupby('Country Name')['Year'].count()
complete_countries = country_years_count[country_years_count == len(years)].index
balanced_panel = wide_form_data[wide_form_data['Country Name'].isin(complete_countries)]
print(balanced_panel)
```

```python
#Question 13: 
#1. Read the CSV file again, import all data as strings. Named your Dataframe "exchange_rate_df"
exchange_rate_df = pd.read_csv("C:/Users/uy128/Desktop/IFS_09-16-2024 16-36-10-58_timeSeries.csv", dtype=str)
#2. Select Years (2010 to 2023)
selected_years_data = exchange_rate_df.loc[:, '2010':'2023']
print(selected_years_data)
```

```python
#Question 13:
#3. Filter data where column "Attribute" contains only the attribute Value
exch_filtered_data = exchange_rate_df[exchange_rate_df['Attribute'] == 'Value']
print(exch_filtered_data)
```

```python
#Question 13:
4. #Filter data where column "Indicator Code" contains only "ENDA_XDC_USD_RATE" which is 
#Exchange Rates, Domestic Currency per U.S. Dollar, Period Average, Rate.

filtered_indicator_data = exch_filtered_data[exch_filtered_data['Indicator Code'] == 'ENDA_XDC_USD_RATE']
print(filtered_indicator_data)
```

```python
#Question 13: 
#5. Remove all columns except "Country Name", and Years (2010 to 2023)
final_data = filtered_indicator_data[['Country Name']].join(selected_years_data)
print(final_data)
```

```python
#Question 13:
#6. Reshape the dataframe from wide to long The years are organized under a new column named
#"Year" , and he values are organized under a new column named "ENDA_XDC_USD_RATE"

long_data = final_data.melt(id_vars=['Country Name'], 
                             var_name='Year', 
                             value_name='ENDA_XDC_USD_RATE')
print(long_data)
```

```python
#Question 13:
#7. Join this dataframe with the data at you prepared at the end of Question 12. 
#Make sure the result drop extra countries from Dataframe "exchange_rate_df" which are not in your main dataframe.
merged_data = pd.merge(long_data, balanced_panel, on=['Country Name', 'Year'], how='inner')
print(merged_data)
```

```python
#Question 14: Calculate
#1. "Real_GDP_USD" using Equation 13
#Equation 13: Real_GDP_USD = (R/ENDA_XDC_USD_RATE)

merged_data['Real_GDP_USD'] = merged_data['R'].astype(float) / merged_data['ENDA_XDC_USD_RATE'].astype(float)
print(merged_data[['Country Name', 'Year', 'Real_GDP_USD']])
```

```python
#Question 14:
#2. "Nominal_GDP_USD" using Equation 14
#Equatoin 14: (N/ENDA_XDC_USD_RATE)

merged_data['Nominal_GDP_USD'] = merged_data['N'].astype(float) / merged_data['ENDA_XDC_USD_RATE'].astype(float)
print(merged_data[['Country Name', 'Year', 'Nominal_GDP_USD']])
```

```python
#Question 14: Calculate
#3. Mean "Real_GDP_USD" and mean "Nominal_GDP_USD" across countries. 
#Order the result in ascending order of mean"Real_GDP_USD"

mean_gdp = merged_data.groupby('Country Name')[['Real_GDP_USD', 'Nominal_GDP_USD']].mean().reset_index()
mean_gdp_sorted = mean_gdp.sort_values(by='Real_GDP_USD', ascending=True).reset_index()
print(mean_gdp_sorted)
```
